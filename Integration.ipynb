```python
#Methode des rectangles
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
  return 3*x**2 + 2*x -1 + np.sin(2*np.pi*x) +np.exp(-x)*np.e/(np.e -1)
```

```python
def rectangles(f,a,b,N):
  delta = (b-a)/N
  J=0
  for i in range(N):
    x = a + delta * i
    J += f(x) * delta
  return J
```

```python
rectangles(f, 0, 1, 100)
```

```python
def rectangles_aplpha(f, a, b, N, alpha=0.5):
  delta = (b-a)/N
  J = 0
  for i in range(N):
    x = a + delta * (i + alpha)
    J += f(x) * delta
  return J
```

```python
rectangles_aplpha(f, 0 , 1, 100, alpha=0.5)
```

```python
def plot_rectangles(f, a, b, N, alpha=0.5):
  plt.figure(figsize=(8,5))
  plt.title(f"Methode des rectangles $\\alpha = {alpha}$ et $N ={N}$")
  #Vectorialize f
  f = np.vectorize(f)
  delta = (b-a)/N
  x = np.linspace(a + alpha + delta, b + alpha * delta, N, endpoint=False)
  y = f(x)
  plt.plot(x,y,"bo-")
  #dessine les boites
  bx = np.linspace(a,b,N+1,endpoint=True)
  for i in range(N):
    x_rect = [bx[i], bx[i], bx[i+1], bx[i+1], bx[i]]
    y_rect = [0, y[i], y[i], 0,0]
    plt.plot(x_rect, y_rect, 'r')
```

```python
plot_rectangles(f,0,1,10,alpha=0.2)
```
