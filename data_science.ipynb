#Gestion de temps

```python
import time
```

```python
time.gmtime(0)
```

```python
def timeit(function, *args, **kwargs):
  t = time.time()
  time.sleep(1)
  function(*args,**kwargs)
  return time.time() -t
```

```python
timeit(sample, range(5), 3)
```

```python
timezones = [
    "Africa/Sao_Tome",
    "America/Los_Angeles",
    "America/New_York",
    "Asia/Hong_Kong",
    "Europe/Paris",
    "Pacific/Noumea",
    "Pacific/Tahiti",
]
```

```python
def saint_sylvestre(tz):
  return datetime(2020, 1, 1, tzinfo=ZoneInfo(tz))
```
