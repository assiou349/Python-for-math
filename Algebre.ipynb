```python
import numpy as np
```

```python
def LU(A):
  #dimensions
  n = A.shape[0]
  #initialisation des matrice L et U
  L = np.identity(n)
  U = np.zeros((n,n))
  #boucle sur i
  for i in range(n):
    #calcule de la i-eme ligne de u
    for j in range(i,n):
      U[i,j] = A[i,j] - np.dot(L[i,:i], U[:i,j])

    if abs(U[i,i]) < 1e-16:
      raise Exception("Pivot null")

    for j in range(i+1,n):
      L[j,i] = (A[j,i] - np.dot(L[j,:i], U[:i,i]))/U[i,i]

  return L, U
```

```python
A = np.random.random((10,10))
L, U = LU(A)
print(np.allclose(A, L @ U))
```

```python
def solve(L,U,b):
  n = L.shape[0]
  #resoud Ly = b
  y = np.zeros(n)
  for i in range(n):
    y[i] = b[i]- np.dot(L[i,:i], y[:i])
  x = np.zeros(n)
  for i in range(n-1, -1, -1):
    x[i] = (y[i] - np.dot(U[i,i+1:],x[i+1:]))/U[i,i]
  return x
```

```python
A = np.random.random((4,4))
b = np.random.random(4)

L, U = LU(A)
res = A @ solve(L, U, b) - b
print(np.linalg.norm(res))
print(np.allclose(A @ solve(L, U, b), b))
```

```python
A@solve(L,U,b)-b
```

```python
def LU_inplace(A):
  # dimensions
  n = A.shape[0]
  for i in range(n):
    #calcule la i-eme ligne de U
    for j in range(i, n):
      A[i,j] = A[i,j]  - np.dot(A[i,:i], A[:i,j])
    if abs(A[i,i]) <1e-16:
      raise Exception("Pivot null")
    for j in range(i+1, n):
      A[j,i] = (A[j,i] -np.dot(A[j,:i], A[:i,i])) /A[i,i]
  return A
```

```python
def solve_inplace(A, b):
  n = A.shape[0]
  #resoud Ly = b in place
  for i in range(n):
    b[i] = b[i] - np.dot(A[i,:i], b[:i])
  #resoud Ux = y
  for i in range(n-1, -1,-1):
    b[i] = (b[i] - np.dot(A[i,i+1:],b[i+1:]))/A[i,i]
  return b
```

```python
A = np.random.random((4,4))
b = np.random.random(4)
Ak = A.copy()
bk = b.copy()

A = LU_inplace(A)
Ak  @ solve_inplace(A,b) - bk
```
