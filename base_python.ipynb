```python
#schema de horner
def horner(elt: str) -> int:
  "Calcule la valer d'un entier en suivant le schema de horner"

  result = 0
  for digit in elt:
    result =  result * 10 + (ord(digit) - ord("0"))
  return result
```

```python
#crible d'eratosthene
def crible_eratosthene(n: int) -> set:
  "Enumere les nombre premier inferieur a n"

  #on cree d'abord la grille complete
  p = set(range(2,n))
  #puis pour chaque entier i
  for i in range(2,n):
    #on elimine l'ensemble des multiples de i
    p = p - set(x*i for x in range(2, n // i +i))
  return p
```

```python
crible_eratosthene(20)
```

#Fonction

```python
def enfants(nom: str, age: int) -> bool:
  "Renvoie True pour les moins de 18 ans"
  return age <= 18
```

```python
def groupe(personnes: dict, condition) -> list:
  """ Renvoie une liste de personnes remplissent une condition.

  L'argument condition est une fonction qui renvoie True
  si l'age respect un certains critere.
  """
  return list(
      nom for (nom, age) in personnes.items()
      if condition(nom, age)
  )
```

```python
tous = {
    "Jules":5, "Marie":17, "Pierre":21, "Jacques":80
}
```

```python
groupe(tous, enfants)
```

```python
groupe(tous, lambda nom, age: age >= 70)
```

```python
groupe(tous, lambda nom, age: nom.startswith("J"))
```

```python
groupe(personnes=tous, condition=enfants)
```

```python
groupe(condition=enfants, personnes=tous)
```

```python
def nouveau_point(
    x: float, y: float = 0, /, z: float = 0, *args, color:str = 'red',
    temperature: float = 25, **kwargs):
  if len(args)>0:
    print(f"Parametre non nomme ignore: {args}")
  if len(kwargs) > 0:
    print(f"Parametre nomme ignore: {kwargs}")

  return{
      'x': x, 'y':y, 'z':z, 'color': color, 'temperature': temperature,
  }
```

```python
nouveau_point(1, temperature=12, color="blue", size=5)
```

```python
def nouveau(fonction, x: float, y: float, *args, **kwargs):
  "Cree un nouvel element defini par 'function"
  return fonction(x,y, *args, **kwargs)
```

```python
def nouveau(fonction, fn_args: tuple, fn_kwargs:dict):
  "Cree un nouvel element defin par 'fonction'."
  return fonction(*fn_args, **fn_kwargs)
```

```python
def new_dict(original_data: dict=dict(), **kwargs) -> dict: #Mauvais pattern, ne pas mettre d'objet mutable par defaut
  for key, value in kwargs.items():
    original_data[key] = value
  return original_data
```

```python
new_dict(color="red", value=1)
```

```python
new_dict(length=1)
```

```python
def update_dict(original_data = None, **kwargs) -> dict:
  if original_data is None:
    original_data = dict()
  for key, value in kwargs.items():
    original_data[key] = value
  return original_data
```

```python
update_dict(color="red", value=1)
```

```python
update_dict(length=12)
```

```python
def pgcd(a: int, b: int) -> int:
  """Calcul du PGCD de deux entiers.

  Les parametres a et b sont convertis en entier avant le calcul du PGCD
  Une exception valueError est levee si un des entier est negatif ou null.
  """
  a,b = int(a), int(b)
  if ( a<=0 or b <=0):
    raise ValueError("a et b doivent etre deux entiers positifs")
  while a != b:
    if a > b:
      a = a-b
    else:
      b = b-a
    return a
```

```python
def print_pgcd(elts: list) -> None:
  for (a, b) in elts:
    try:
      x = pgcd(a, b)
    except ValueError as e:
      #traitement de l'exception valueError
      print(f"ValueError: {e}")
    else:
      print(f"pgcd({a},{b}) = {x}")
    finally:
      #execute dans tous les cas
      print("----")
```

```python
print_pgcd([(12,8),(7,"a"),(4,2.3),("42", 56)])
```
