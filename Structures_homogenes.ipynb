```python
def diff_list(lst):
  return [lst[i+1] - lst[i] for i in range(len(lst) -1)]
```

```python
def diff_list(lst):
  return [b -a for a, b in zip(lst, lst[1:])]
```

```python
def dif_np(vec):
  return vec[1:] - vec[:-1]
```

```python
#Matrice de Vandermonde
def vdm_1(x, n):
  p =  np.size(x)
  vdm = np.empty((p, n+1))
  for i in range(p):
    for j in range(n+1):
      vdm[i,j] = x[i]**j
  return vdm
```

```python
def vdm_2(x,n):
  p = np.size(x)
  vdm = np.empty((p, n+1))
  for k in range(n+1):
    vdm[:,k] = x**k
  return vdm
```

```python
def vdm_3(x, n):
  p = np.size(x)
  vdm = np.ones((p,n+1))
  for j in range(1, n+1):
    vdm[:,j] = vdm[:,j-1]*x
  return vdm
```
