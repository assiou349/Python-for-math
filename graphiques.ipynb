```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,1,50)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title(r'Graphique de $x^2$')
plt.xlabel(r'$X$')
plt.ylabel(r'$Y$')
plt.plot(x, y, marker='o',label=r"$x^2$")
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(8,5))
plt.title(r'Fonctions trigonometriques')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$')
#plt.xticks([0, np.pi /2, np.pi, 3 * np.pi/2, 2 * np.pi],[r'$0$', r'$np.pi /2$', r'$np.pi$', r'$3 * np.pi/2$', r'$2 * np.pi$'])
for k in range(1,4):
  plt.plot(x, np.sin(k*x))
  plt.plot(x, np.cos(k*x))
plt.legend()

```
